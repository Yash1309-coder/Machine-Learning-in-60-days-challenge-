#List


```python
#list should be include in the square brackets
my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
# list can have multiple data types
my_list = [2,3,1.8, "English", True]
print(my_list)
```

```python
#add an elemnt to a lsit
my_list = [2,3,1.8, "English", True]
my_list.append(6)
print(my_list)
```

```python
# print elements of a list using their index
print(my_list[0])
print(my_list[2])
```

```python
# list allow duplicate values
list_l = [1,2,3,4,5,6,77,2,3,]
print(list_l)
```

```python
print(len(list_l))
```

```python
# delete an item in a list
del list_l[2]
print(list_l)
```

```python
# join 2 lists
list_3 = my_list + list_l
print(list_3)
```

# Tuple


```python
tuple_1 = (2,3,4,5)
print(tuple_1)
type(tuple_1)
```

```python
# tuple allows multiple data types
tuple_2 = (1,2,3,5, "Machine Learning", False)
print(tuple_2)
```

```python
#converting a list to a tuples
my_list = [3,4,5,6]
print(my_list)

my_tuple = tuple(my_list)
print(my_tuple)
type(my_tuple)
```

```python
print(my_tuple[0])
print(my_tuple[3])
```

```python
print(len(my_tuple))
```

#Set


```python
# set --> Curly brackets
my_set = {1,2,3,4,5}
print(my_set)
type(my_set)
```

```python
# set does not support indexing
print(my_set[0])
```

```python
# convert a list to a set
list_5 = [4,5,6,7,8]
x = set(list_5)
print(x)
```

```python
# set does not allow duplicate values
set_3 = {1,2,3,4,5,5,6,7}
print(set_3)
```

#Dictionary

key-Value Pair

```python
my_dictionary = {'name' : 'David', 'age' : 30 , 'country' : 'India'}
print(my_dictionary)
type(my_dictionary)
```

```python
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['country'])
```

```python
# dictionary does not allow duplicate values

dictionary_2 = {'name' : 'David', 'age' : 30 , 'country' : 'India', 'name' : 'David', 'age' : 30 , 'country' : 'India'}
print(dictionary_2)
```
