1. Arithmetic Operators


```python
num_1 = 20
num_2 = 10

# addition
sum = num_1 + num_2
print('sum = ' , sum )

# subtraction
diff = num_1 - num_2
print('difference = ' , diff)

# multiplication
pro = num_1 * num_2
print('product = ', pro)

# division 
quo = num_1 / num_2
print('quotient = ' , quo)

# exponent 
exp = num_1**num_2
print('exponent = ', exp )

# modulus
mod = num_1 % num_2
print('reminder = ' , mod)
```

3. Comparision Operators

```python
a = 5
b = 10

print(a == b) # equal to
print(a != b) # not equal to
print( a > b ) # greater then
print( a < b )
print( a >= b )
print( a <= b )
```

```python
a = 10

print(a>20 and  a>5)
print(a>20 or a>5)
print(not(a>8 and a>5))
```

5. Identity Operator

```python
x = 5
y = 5
print(x is y)
```

6. Membership Operators

in

not in 

```python
a = 5
b = 10
c = [1,2,3,4,5]
print(a in c)
print(b in c)
```

```python
a = 5
b = 10
c = [1,2,3,4,5]
print(a not in c)
print(b not in c)
```
